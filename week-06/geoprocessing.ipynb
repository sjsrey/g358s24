## Dissolve

```python
counties = gdf1.dissolve(by='county')
```

```python
counties.plot()
```

```python
counties.shape
```

```python
gdf1.shape
```

```python
county_centroids = counties.centroid
```

```python
base = counties.plot()
county_centroids.plot(color='r', ax=base)
```

```python
counties.crs = gdf1.crs
```

```python
counties.crs
```

```python
counties.area
```

```python
buffer = county_centroids.buffer(16093.4) # 10 mile buffer
```

```python
base = counties.plot()
buffer.plot(color='green', ax=base)
county_centroids.plot(color='r', ax=base)
```

```python
gdf1.sindex.query(buffer, predicate='intersects')
```

```python
res = gdf1.sindex.query(buffer, predicate='intersects')
```

```python
gdf1.iloc[res[1]].plot()
```

```python
base = gdf1.plot(alpha=0.5)
gdf1.iloc[res[1]].plot(ax=base, color='yellow')
buffer.plot(ax=base, color='green')
```

```python
counties.boundary
```

```python
counties.boundary.plot()
```

```python
bbuffer = counties.boundary.buffer(10000)
```

```python
bbuffer.plot()
```

```python
ints = gdf1.sindex.query(bbuffer, predicate='intersects')
```

```python
ints.shape
```

```python
gdf1.iloc[ints[1]].plot()
```

```python
gdf1.iloc[~ints[1]].plot()
```

```python
base = gdf1.plot()
gdf1.iloc[ints[1]].plot(color='y', ax=base)
```

```python
import numpy as np
boundary = np.array(['False'] * gdf1.shape[0])
boundary[ints[1]]=True
gdf1['Boundary'] = boundary
```

```python
gdf1.plot(column='Boundary', categorical=True, legend=True)
```
